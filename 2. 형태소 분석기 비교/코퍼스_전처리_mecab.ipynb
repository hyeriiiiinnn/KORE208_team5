```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', text)                 # HTML 태그 제거
    text = re.sub(r'\n+', ' ', text)                  # 줄바꿈 -> 공백
    text = re.sub(r'["“”‘’\'!?…·,;:\(\)\[\]\{\}]', '', text)  # 주요 문장부호 제거
    text = re.sub(r'[^\w\s가-힣]', '', text)           # 그 외 특수문자 제거 (이중 필터링)
    text = re.sub(r'\s+', ' ', text).strip()          # 다중 공백 제거
    return text

# 적용
cleaned_contents = [clean_text(text) for text in contents]

```

```python

# 4. 문장 단위로 품사 태깅 결과를 병합 (Mecab 사용)
import re
import pandas as pd

def split_sentences(text):
    return [s.strip() for s in text.split("\n") if s.strip()]

# 1) tagged_articles 생성
tagged_articles = []
for doc_id, text in enumerate(contents):
    article_sents = []
    for sent in split_sentences(text):
        # (token, tag) 튜플 리스트
        tagged_tokens = mecab.pos(sent)
        article_sents.append({
            "doc_id": doc_id,
            "sentence": sent,
            "tagged_tokens": tagged_tokens
        })
    tagged_articles.append(article_sents)

# 2) DataFrame 변환
rows = []
for article in tagged_articles:
    for sent in article:
        for token, tag in sent["tagged_tokens"]:
            rows.append({
                "doc_id": sent["doc_id"],
                "sentence": sent["sentence"],
                "token": token,
                "tag": tag
            })

df = pd.DataFrame(rows)

```
